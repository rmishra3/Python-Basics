```python
import cv2
import os
```

```python
img_path=os.path.join('Basic_Data','Hand1.jpg')
img=cv2.imread(img_path)
img
```

```python
!pip install matplotlib
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
plt.show
```

```python
recolor_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(recolor_img)
plt.imshow(gray_img)
plt.show()
```

```python
output_path=os.path.join('Basic_Data','Gray_Hand1.jpeg')
cv2.imwrite(output_path,gray_img)
```

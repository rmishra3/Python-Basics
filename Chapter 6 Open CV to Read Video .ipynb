```python
# Importing all the required dependencies
import cv2
import os
from matplotlib import pyplot as plt
```

```python
# Making connection with the video. It can be  ppprerecorded videos, USB connected camera or microscope
cap=cv2.VideoCapture(os.path.join('Basic_Data','Videos','Res1.MOV'))
# Reading one frame of the video. ret tell us if the frame was successfully read. Frame is the image.
ret,frame=cap.read()
frame.shape
frame1=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(frame1)
plt.show()
```

```python
# Grabbing video properties
Img_Height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
Img_Width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
Img_Count=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
Img_FPS=int(cap.get(cv2.CAP_PROP_FPS))

print('Height={}, Width= {}, Count={}, Frame Rate={} and Video Duration = {} seconds'.format(Img_Height,Img_Width,Img_Count,Img_FPS,Img_Count/Img_FPS))
```

```python
# Displaying video --> Modify it --> Save the video

cap=cv2.VideoCapture(os.path.join('Basic_Data','Videos','Res1.MOV'))

Video_Writer = cv2.VideoWriter(os.path.join('Basic_Data','Videos','Output_Gray.avi'),cv2.VideoWriter_fourcc('P','I','M','1'), Img_FPS, (Img_Width, Img_Height), isColor=False)
#video_writer = cv2.VideoWriter(os.path.join('data','output','output.avi'), cv2.VideoWriter_fourcc('P','I','M','1'), fps, (width, height), isColor=False) 
for Frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    ret,Frame1=cap.read()
    Frame=cv2.cvtColor(Frame1,cv2.COLOR_BGR2GRAY) 
    cv2.imshow('Surgery',Frame)
     # Breaking out of the loop
    Video_Writer.write(Frame)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break  
cap.release()
cv2.destroyAllWindows()
```
